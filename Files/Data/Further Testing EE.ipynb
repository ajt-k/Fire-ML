```python
year = 2017
reference_date = ee.Date('2017-01-01')

def compute_bboxes(feature):

    geometry = feature.geometry().bounds()
    start_date = reference_date.advance(feature.get('min'),'day')
    end_date = reference_date.advance(feature.get('max'),'day')

    mod_feature = feature.setGeometry(geometry).set({'start_date': start_date, 'end_date': end_date})
    
    return  mod_feature


bboxed_features = features.map(lambda feature: compute_bboxes(feature))

date = ee.Date(bboxed_features.first().get('start_date'))
print(date.format().getInfo())

print(ee.Feature(bboxed_features.first()).id().getInfo())
```

```python
x_data_wind = ee.ImageCollection("ECMWF/ERA5_LAND/HOURLY").select('u_component_of_wind_10m').set('description', None)
y_data_wind =  ee.ImageCollection("ECMWF/ERA5_LAND/HOURLY").select('v_component_of_wind_10m').set('description', None)




def clip_collection(feature):
    collection = x_data_wind
    start_date = feature.get('start_date')
    end_date = feature.get('end_date')
    geometry = feature.geometry()
    id = feature.get('system:id')
    
    clipped_images = collection.filterDate(start_date, end_date).filterBounds(geometry).map(lambda img: img.clip(geometry))

    
    # Convert ImageCollection to a list
    clipped_images_list = clipped_images.toList(clipped_images.size())
    
    # Create a dictionary with metadata
    metadata = {
        'id': id,
        'start_date': start_date,
        'end_date': end_date,
        'geometry': geometry,
        'clipped_images': clipped_images_list
    }
    
    return ee.Feature(None, metadata)





clipped = bboxed_features.map(clip_collection)

pprint.pprint(clipped.first().propertyNames().getInfo())



```

```python
import ee

# Authenticate to Earth Engine
ee.Authenticate()
ee.Initialize()

# Access the monthly burned pixel product dataset
fire_dataset = ee.ImageCollection('ESA/CCI/FireCCI/5_1')

# Filter the dataset by time (e.g., a specific year)
filtered_dataset = fire_dataset.filterDate('2022-01-01', '2022-12-31')

# Apply a threshold to identify fire-affected areas
thresholded_dataset = filtered_dataset.select('BurnDate').gte(1)  # Example threshold

# Detect individual fire instances using connectedPixelCount()
fire_instances = thresholded_dataset.connectedPixelCount(256, False)

# Define a function to extract fire properties
def extract_fire_properties(feature):
    date = ee.Date(feature.get('system:time_start')).format("YYYY-MM-dd")
    centroid = feature.geometry().centroid()
    return feature.set({'date': date, 'centroid': centroid})

# Reduce connected components to extract properties of each fire instance
fire_properties = fire_instances.reduceConnectedComponents(
    reducer=ee.Reducer.mean(),  # This reducer doesn't change the pixel values, only computes statistics
    labelBand='labels',         # The band containing connected component labels
    reducerOptions={'geometry': 'centroid', 'reducer': ee.Reducer.first()})
  
# Map the function over each fire instance and extract its properties
fire_features = fire_properties.map(extract_fire_properties)

# Extract date, time, latitude, and longitude of each fire instance
fire_data = fire_features.aggregate_array('date').zip(
    fire_features.aggregate_array('centroid').map(lambda point: point.coordinates()))

# Print the result
print('Date, Latitude, Longitude:')
print(fire_data.getInfo())
```
